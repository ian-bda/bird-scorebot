```python
import pandas as pd
```

```python
# File 1: has columns: Species, Code, North America
df1 = pd.read_csv("/Users/ibirc/Downloads/Bird_Rankings - Sheet1.csv")

# File 2: has columns: Species, North America
df2 = pd.read_csv("/Users/ibirc/Downloads/species_rarity_by_continent_yearly - species_rarity_by_continent_yearly.csv")

# Show first few rows to check
print("=== File 1 ===")
display(df1.head())

print("=== File 2 ===")
display(df2.head())
```

```python
# First, set index on Species so update works easily
df1_indexed = df1.set_index('Species')
df2_indexed = df2.set_index('Species')

# Replace North America column in df2 with values from df1 (for matching Species)
df2_indexed.update(df1_indexed[['North America']])

# Reset index to get Species back as a column
df2_updated = df2_indexed.reset_index()

# Convert 'North America' column to integer, then back to string if you want no decimals
df2_updated['North America'] = df2_updated['North America'].astype(int).astype(str)


display(df2_updated.head())
```

```python
# Save
df2_updated.to_csv('/Users/ibirc/Downloads/file2_updated.csv', index=False)
print("Updated file saved as 'file2_updated.csv'")
```
